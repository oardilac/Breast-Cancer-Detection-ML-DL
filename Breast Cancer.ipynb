## Section 3: Data preparation for modeling

```python
# We separate the independent and dependent variables
X = df.drop('diagnosis_M', axis=1)
y = df['diagnosis_M']
```

```python
# Standardize data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

# We apply PCA
n_componentes = range(1, 31)
pca = PCA(n_components=30)
pca.fit(X)
```

```python
var_exp_cumulative = np.cumsum(pca.explained_variance_ratio_)
plt.plot(n_componentes, var_exp_cumulative)
plt.xlabel('Numero de componentes')
plt.ylabel('Varianza acumulada')
plt.show()
```

```python
pca_95 = PCA(n_components=0.95, random_state=2020)
pca_95.fit(X)
X = pca_95.transform(X)
```

```python
X.shape
```

```python
df_pca = pd.DataFrame(X, columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10'])
df_pca['diagnosis'] = y
df_pca.head()
```

```python
X_new = df_pca.drop(['diagnosis'], axis=1)
y_new = df_pca['diagnosis']
```

## Section 4: Modeling and evaluation

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm = SVC(kernel='linear', probability=True)
svm.fit(X_train, y_train)

# Making predictions
y_pred = svm.predict(X_test)
print('Precisión:', accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel('Predicciones')
plt.ylabel('Valores reales')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier

# Define the models
dtc = DecisionTreeClassifier()
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
knn = KNeighborsClassifier()

# Add them to the VotingClassifier
votacion_dura = VotingClassifier(
    estimators=[
        ('svm', svm), 
        ('tree', dtc), 
        ('random_forest', rf), 
        ('gradient_boosting', gb), 
        ('knn', knn)
    ], 
    voting='hard'
)

# Training the VotingClassifier
votacion_dura.fit(X_train, y_train)

# Making predictions
y_pred_hard = votacion_dura.predict(X_test)
print('Precisión:', accuracy_score(y_test, y_pred_hard))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_hard)
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel('Predicciones')
plt.ylabel('Valores reales')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameters and their values for SVC
param_grid_svc = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 

grid_svc = GridSearchCV(SVC(), param_grid_svc, refit=True, verbose=3)

# Fit the grid search object to the data
grid_svc.fit(X_train, y_train)

# print best parameter after tuning 
print(grid_svc.best_params_) 

# print how our model looks after hyper-parameter tuning 
print(grid_svc.best_estimator_)
```

```python
# Define the hyperparameters and their values for DecisionTreeClassifier
param_grid_dtc = {'criterion': ['gini', 'entropy'], 
              'max_depth': [2, 4, 6, 8, 10, None],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 5, 10]} 

grid_dtc = GridSearchCV(DecisionTreeClassifier(), param_grid_dtc, refit=True, verbose=3)

# Fit the grid search object to the data
grid_dtc.fit(X_train, y_train)

# Print best parameters after tuning 
print(grid_dtc.best_params_)

# Print model after hyper-parameter tuning 
print(grid_dtc.best_estimator_)
```

```python
# Define the hyperparameters and their values for RandomForestClassifier
param_grid_rf = {'n_estimators': [10, 100, 200, 500],
              'max_features': ['auto', 'sqrt'],
              'max_depth' : [4,5,6,7,8],
              'criterion' :['gini', 'entropy']}

grid_rf = GridSearchCV(RandomForestClassifier(), param_grid_rf, refit=True, verbose=3)

# Fit the grid search object to the data
grid_rf.fit(X_train, y_train)

# Print best parameters after tuning 
print(grid_rf.best_params_)

# Print model after hyper-parameter tuning 
print(grid_rf.best_estimator_)
```

```python
# Define the hyperparameters and their values for KNeighborsClassifier
param_grid_knn = {'n_neighbors': [3, 5, 11, 19],
              'weights': ['uniform', 'distance'],
              'metric': ['euclidean', 'manhattan']}

grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, refit=True, verbose=3)

# Fit the grid search object to the data
grid_knn.fit(X_train, y_train)

# Print best parameters after tuning 
print(grid_knn.best_params_)

# Print model after hyper-parameter tuning 
print(grid_knn.best_estimator_)
```

```python
# Predicting with the best models obtained after hyper-parameter tuning
models = [grid_svc, grid_dtc, grid_rf, grid_knn]

for model in models:
    y_pred = model.predict(X_test)
    print('Model:', model)
    print('Accuracy:', accuracy_score(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, cmap='Blues')
    plt.xlabel('Predictions')
    plt.ylabel('Actuals')
    plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Define the models and parameters
model_params = {
    'svm': {
        'model': SVC(),
        'params': {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {'criterion': ['gini', 'entropy'], 'max_depth': [2, 4, 6, 8, 10, None],
                   'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 5, 10]}
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {'n_estimators': [10, 100, 200, 500], 'max_features': ['auto', 'sqrt'],
                   'max_depth' : [4,5,6,7,8], 'criterion' :['gini', 'entropy']}
    },
    'knn': {
        'model': KNeighborsClassifier(),
        'params': {'n_neighbors': [3, 5, 11, 19], 'weights': ['uniform', 'distance'], 
                   'metric': ['euclidean', 'manhattan']}
    }
}

# Perform GridSearchCV for each model
from sklearn.model_selection import GridSearchCV

scores = []

for model_name, mp in model_params.items():
    grid = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    grid.fit(X_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': grid.best_score_,
        'best_params': grid.best_params_
    })
    
df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Get the number of features
n_features = X_train.shape[1]

# Define the model
model = Sequential()
model.add(Dense(32, activation='relu', kernel_initializer='he_normal', input_shape=(n_features,)))
model.add(Dense(16, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Fit the model
model.fit(X_train, y_train, epochs=150, batch_size=32, verbose=2, validation_split=0.2)

# Evaluate the model
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print('Test Accuracy: %.3f' % acc)
```
